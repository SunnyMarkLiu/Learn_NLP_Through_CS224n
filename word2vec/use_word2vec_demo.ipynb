## Using pre-trained Word2Vec model

```python
class MySentences(object):
    def __init__(self, dirname):
        self.dirname = dirname
 
    def __iter__(self):
        for fname in os.listdir(self.dirname):
            for line in open(os.path.join(self.dirname, fname)):
                """
                convert to unicode, lowercase, remove numbers, extract named entities… 
                """
                yield line.split()

more_sentences = MySentences('/some/directory') # a memory-friendly iterator
```

```python
model.train(sentences=more_sentences, epochs=5)
```

```python
model.save('/new/model/path')
```
